# Convolution Filters
This is a useful notebook to look at convolution filters with OpenCV.

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2                              # pip install opencv-python
from skimage.data import astronaut      # pip install scikit-image
```

```python
def show_gray_image(im):
    frame = plt.gca()
    frame.axes.get_xaxis().set_visible(False)
    frame.axes.get_yaxis().set_visible(False)
    plt.imshow(im, cmap='gray')
```

## Astronaut Image
We are going to use the astronaut image from scikit-image as the starting point for our convolution filters.

```python
image = astronaut()
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # convert the color image to gray
show_gray_image(gray)
```

## Laplacian

```python
laplacian = np.array(([0, 1, 0], 
                      [1, -4, 1], 
                      [0, 1, 0]), dtype="int")
laplacian_img = cv2.filter2D(gray, -1, laplacian)
show_gray_image(laplacian_img)
```

## Mean blur

```python
mean_blur = np.ones((3,3), dtype="float") * (1.0/9)
mean_blur_img = cv2.filter2D(gray, -1, mean_blur)
show_gray_image(mean_blur_img)
```

## Gaussian blur

```python
gaussian_blur = np.array(([1, 2, 1], 
                          [2, 4, 2], 
                          [1, 2, 1]), dtype="float") * (1.0/16)
gaussian_blur_img = cv2.filter2D(gray, -1, gaussian_blur)
show_gray_image(gaussian_blur_img)
```

## Sharpen

```python
sharpen = np.array(([0, -1, 0], 
                    [-1, 5, -1], 
                    [0, -1, 0]), dtype="int")
sharpen_img = cv2.filter2D(gray, -1, sharpen)
show_gray_image(sharpen_img)
```

## Sobel filters
We are going to look at the horizontal and vertical Sobel filters.

```python
sobel_horizontal = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)
show_gray_image(sobel_horizontal)
```

```python
sobel_vertical = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)
show_gray_image(sobel_vertical)
```
